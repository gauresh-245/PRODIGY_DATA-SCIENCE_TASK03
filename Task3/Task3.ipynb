```python
high_corr_cols = ['emp.var.rate', 'euribor3m', 'nr.employed']
```

```python
df1=df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols,inplace=True,axis=1) #axis=1 indicates columns;
df1.columns
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
lb = LabelEncoder()

# Apply LabelEncoder to categorical columns only
df_encoded = df1.apply(lambda col: lb.fit_transform(col) if col.dtype == 'object' else col)

# Print the encoded DataFrame to verify
print(df_encoded)
```

```python
df_encoded['deposit'].value_counts()
```

```python
x=df_encoded.drop('deposit',axis=1)  # indepedent variable
y=df_encoded['deposit'] # depedent variable
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
    acc =  accuracy_score(y_test,y_pred)
    print('Accuracy_score',acc)
    cm=confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('classification Report\n',classification_report(y_test,y_pred))

def mscore(model):
    train_score=model.score(x_train,y_train)
    test_score= model.score(x_test,y_test)
    print('training score',train_score)
    print('Testing score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt= DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train)
```

```python
def mscore(modeel):
    train_score = modeel.score(x_train, y_train)
    test_score = modeel.score(x_test, y_test)
    print('training score:', train_score)
    print('test score:', test_score)
```

```python
mscore(dt)
```

```python
ypred_dt=dt.predict(x_test)
print(ypred_dt)
```

```python
eval_model(y_test,ypred_dt)
```
